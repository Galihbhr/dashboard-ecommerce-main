### Assessing Data

```python
def inspect_data(df):
    # Menampilkan ringkasan informasi data
    print("Informasi DataFrame:")
    print(df.info(), "\n")

    # Menampilkan jumlah nilai null di setiap kolom
    print("Jumlah Nilai Null per Kolom:")
    print(df.isnull().sum(), "\n")

    # Menampilkan ringkasan statistik deskriptif
    print("Deskripsi Statistik:")
    print(df.describe(), "\n")

    # Menampilkan jumlah duplikasi data
    print("Jumlah Duplikasi:")
    print(df.duplicated().sum(), "\n")
```

## Visualization & Explanatory Analysis

```python
product_order_df = pd.merge(order_items_data,products_data,on = 'product_id')
product_order_df = pd.merge(product_order_df,product_category_name_translation, on = 'product_category_name')
```

```python
all_df = pd.merge(orders_data, customers_data, on="customer_id", how = 'left')
all_df = pd.merge(all_df, product_order_df, on="order_id", how = "right")
all_df = pd.merge(all_df, sellers_data, on="seller_id", how = 'left')
all_df = pd.merge(all_df, order_payments_data, on="order_id", how = 'left')
all_df = pd.merge(all_df, order_reviews_data, on="order_id", how = 'left')


```

### Pertanyaan 1: pada tahun berapa terjadi penjualan tertinggi?

```python
all_df['year_month'] = all_df['order_purchase_timestamp'].dt.to_period('M')

# Mengelompokkan data berdasarkan tahun dan bulan, menghitung jumlah entri per kombinasi
monthly_counts = all_df.groupby('year_month')['order_item_id'].count()
monthly_counts_df = monthly_counts.reset_index()
monthly_counts_df.columns = ['year_month', 'order_item_id']

# Pastikan kolom 'year_month' diubah ke format datetime
monthly_counts_df['year_month'] = pd.to_datetime(monthly_counts_df['year_month'].astype(str))

# Menampilkan hasil per bulan
print("Jumlah entri per tahun dan bulan:")
print(monthly_counts_df)
```

```python
# Membuat plot garis atau batang dengan sumbu-x datetime
plt.figure(figsize=(14, 6))
sns.barplot(x='year_month', y='order_item_id', data=monthly_counts_df)

# Menyesuaikan label dan judul plot
plt.xlabel('Tahun-Bulan')
plt.ylabel('Jumlah Entri')
plt.title('Distribusi Jumlah Entri per Tahun dan Bulan')
plt.xticks(rotation=45)  # Memiringkan label sumbu-x agar lebih mudah dibaca
plt.tight_layout()  # Menyesuaikan tata letak agar elemen plot tidak terpotong
plt.show()
```

### Pertanyaan 2:

Produk apa yang paling banyak dan paling sedikit terjual?

```python
product_sold_df = all_df.groupby('product_category_name_english')['order_item_id'].sum().sort_values(ascending=False)
product_sold_df.head(10)
```

```python
top_10_products = product_sold_df.head(10)
# Buat visualisasi menggunakan Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(
    x=top_10_products.values,
    y=top_10_products.index,
    palette="viridis"
)

# Menambahkan judul dan label
plt.title("10 Kategori Produk Teratas", fontsize=16)
plt.xlabel("Jumlah Entri", fontsize=12)
plt.ylabel("Kategori Produk", fontsize=12)
plt.tight_layout()

plt.show()
```
